Practice Question : 1. Declare two integer variables, perform all arithmetic operations, and print the results


```python
num1 = 10
num2 = 5

# Perform arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2  # Result will be a float
modulo = num1 % num2
exponentiation = num1 ** num2

# Print the results
print(f"Number 1: {num1}")
print(f"Number 2: {num2}")
print(f"Addition: {addition}")         # 10 + 5 = 15
print(f"Subtraction: {subtraction}")      # 10 - 5 = 5
print(f"Multiplication: {multiplication}")  # 10 * 5 = 50
print(f"Division: {division}")          # 10 / 5 = 2.0
print(f"Modulo: {modulo}")            # 10 % 5 = 0 (remainder)
print(f"Exponentiation: {exponentiation}")  # 10 ** 5 = 100000
```

Write a program that takes user input for age and prints whether the user is an adult (age >= 18).

```python
Age = int(input("Please enter your age: "))

if Age >= 18:
    print("You are an adult.")
else:
    print("You are not yet an adult.")
```
